```python
import heapq as hp
import random
import time
import timeit
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
def tempo_para_heap(fila):
    ini = time.time()
    hp.heapify(fila)
    fim = time.time()
    #print(fila)
    ttotal = fim - ini
    return ttotal

qtd_dados = int(input())
n1 = 1
n2 = 0
contador = 0
Lista_Listas = []
while contador < qtd_dados:
    #nun_itens = int(input())
    n = 0
    fila = []
    for n in range(n2):
        fila.append(random.randint(0, 10))
    #print("Fila antes da Heap:", fila)
    Tempo_total = round(tempo_para_heap(fila), 5)

    Data_line = [n2, Tempo_total]
    Lista_Listas.append(Data_line)

    contador += 1
    n2 += 1
df = pd.DataFrame(Lista_Listas, columns=['itens', 'tempo'])
df.to_excel("planilha_heap.xlsm")
```

```python
class PriorityQueueUnordered:
    def __init__(self):
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0

    def insert(self, item):
        self.queue.append(item)  # Insere o item no final da lista

    def remove(self):
        if self.is_empty():
            raise IndexError("A fila de prioridade est vazia!")

        # Encontra o elemento de maior prioridade
        max_index = 0
        for i in range(len(self.queue)):
            if self.queue[i] > self.queue[max_index]:  # Maior elemento tem maior prioridade
                max_index = i

        # Remove e retorna o elemento de maior prioridade
        return self.queue.pop(max_index)


# Exemplo de uso:
pq = PriorityQueueUnordered()
qtd_dados
cont = 0
p1 = 0
lista = []
definitiva = []
while cont < qtd_dados:
    p2 = 0
    inicio = time.time()
    if p1 != 0:
        for p2 in range(p1):
            pq.insert(random.randint(0, 10))


    fim = time.time()
    ttotal = round(fim - inicio, 5)
    temp = pq.queue
    lista.append(pq.queue)
    temp.clear
    pq.queue = []
    rec = [p1, ttotal]
    definitiva.append(rec)

    cont += 1
    p1 += 1


ef = pd.DataFrame(definitiva, columns=['itens', 'tempo'])
ef.to_excel("planilha_fila_desordenada.xlsm")
```
